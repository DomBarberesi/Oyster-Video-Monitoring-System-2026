```python
from ultralytics import YOLO
import pandas as pd
import os
```

```python
#Validate on 2025 validation set

model_path = "/home/reuekennedy/Oyster-Video-Monitoring-System-2026/YOLO_testing/2025_model/runs/detect/train/weights/best.pt"
dataset25 = "/home/reuekennedy/Oyster-Video-Monitoring-System-2026/YOLO_testing/2025_model/REU_Oyster_2024_Improved-1/data.yaml"
results = []
model = YOLO(model_path)

result = model.val(data=dataset25)
results.append({"Run":"Validation Set 2025", "Precision":round(result.box.p.mean().item(), 3),
    "Recall":round(result.box.r.mean().item(), 3), "mAP@0.5":round(result.box.map50.mean().item(), 3)})
```

```python
#Run on 2025 test set

result = model.val(data=dataset25, split="test")
results.append({"Run":"Test Set 2025", "Precision":round(result.box.p.mean().item(), 3),
    "Recall":round(result.box.r.mean().item(), 3), "mAP@0.5":round(result.box.map50.mean().item(), 3)})
```

```python
from ultralytics import YOLO
import pandas as pd
```

```python
model_path = "/home/reuekennedy/Oyster-Video-Monitoring-System-2026/YOLO_testing/2026_model/runs/detect/models/yolov8s_2026/weights/best.pt"
dataset = "/home/reuekennedy/Oyster-Video-Monitoring-System-2026/YOLO_testing/2026_model/2026-Oyster-Dataset-1/data.yaml"
results = []
model = YOLO(model_path)

result = model.val(data=dataset, conf=0.25)
results.append({"Run":"yolov8s Val", "Precision":round(result.box.p.mean().item(), 3),
    "Recall":round(result.box.r.mean().item(), 3), "mAP@0.5":round(result.box.map50.mean().item(), 3)})
```

```python
result = model.val(data=dataset, split="test", conf=0.25)
results.append({"Run":"yolov8s Test", "Precision":round(result.box.p.mean().item(), 3),
    "Recall":round(result.box.r.mean().item(), 3), "mAP@0.5":round(result.box.map50.mean().item(), 3)})
```

```python
from ultralytics import YOLO
import pandas as pd
```

Run models on validation and test sets

```python
models = ["yolov8s_2026", "yolov8m_2026", "yolov8n_2026",
        "yolov8s_2026_v2", "yolov8m_2026_v2", "yolov8n_2026_v2",
        "yolov8s_2026_v3", "yolov8m_2026_v3",
        "yolov8s_2026_v4", "yolov8m_2026_v4",
        "yolov8s_2026_v5"]
results = []
```

```python
dataset = "/home/reuekennedy/Oyster-Video-Monitoring-System-2026/YOLO_testing/2026_model/2026-Oyster-Dataset-2/data.yaml"
```

```python
for i in range(6):
    model_path = f"/home/reuekennedy/Oyster-Video-Monitoring-System-2026/YOLO_testing/2026_model/runs/detect/models/{models[i]}/weights/best.pt"
    model = YOLO(model_path)
    
    result = model.val(data=dataset, conf=0.25)
    results.append({"Run":f"{models[i]} Val", "Precision":round(result.box.p.mean().item(), 3),
        "Recall":round(result.box.r.mean().item(), 3), "mAP@0.5":round(result.box.map50.mean().item(), 3)})
    
    result = model.val(data=dataset, split="test", conf=0.25)
    results.append({"Run":f"{models[i]} Test", "Precision":round(result.box.p.mean().item(), 3),
        "Recall":round(result.box.r.mean().item(), 3), "mAP@0.5":round(result.box.map50.mean().item(), 3)})
```

```python
dataset = "/home/reuekennedy/Oyster-Video-Monitoring-System-2026/YOLO_testing/2026_model/Oyster-Dataset-2026-1/data.yaml"
```

```python
for i in range(6, 11):
    model_path = f"/home/reuekennedy/Oyster-Video-Monitoring-System-2026/YOLO_testing/2026_model/runs/detect/models/{models[i]}/weights/best.pt"
    model = YOLO(model_path)
    
    result = model.val(data=dataset, conf=0.25)
    results.append({"Run":f"{models[i]} Val", "Precision":round(result.box.p.mean().item(), 3),
        "Recall":round(result.box.r.mean().item(), 3), "mAP@0.5":round(result.box.map50.mean().item(), 3)})
    
    result = model.val(data=dataset, split="test", conf=0.25)
    results.append({"Run":f"{models[i]} Test", "Precision":round(result.box.p.mean().item(), 3),
        "Recall":round(result.box.r.mean().item(), 3), "mAP@0.5":round(result.box.map50.mean().item(), 3)})
```

Compare all model metrics

```python
df = pd.DataFrame(results)
df_sorted = df.sort_values(by="mAP@0.5", ascending=False).reset_index(drop=True)
print("\n2025 Model Metrics")
print(df_sorted)
```
